## Dividing data in features and vectors:

```python
x = winequality_datanew.iloc[:,:-1]
y = winequality_datanew.iloc[:,-1]
```

```python
x
```

```python
y
```

# Transforming data to remove skewness:

```python
from sklearn.preprocessing import power_transform
x = power_transform(x,method = 'yeo-johnson')
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
y = winequality_datanew['quality'].apply(lambda y_value : 1 if y_value>= 7 else 0)
```

## since the outcome has only 2 variables we can use Binary Classification model:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
from sklearn.metrics import accuracy_score
```

# Finding the best random_state:

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = i, test_size = 0.20)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print("At random state",i,"The model performs very well")
        print("random state:-", i)
        print("Training r2_score is:-", accuracy_score(y_train,pred_train)*100)
        print("Testing r2_score is:-", accuracy_score(y_test,pred_test)*100)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = 980)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_test))
```

# Observation: here we can see that the accuracy is 0.79 ie., 79% that means our model predict 79% correct and

# Cross validation Score for logistic regression: 

```python
pred_lr = lr.predict(x_test)
from sklearn.model_selection import cross_val_score
lss = accuracy_score(y_test,pred_lr)
for j in range(2,10):
    lsscore = cross_val_score(lr,x,y,cv = j)
    lsc = lsscore.mean()
    print("At cv :-", j)
    print("Cross validation score is :-" ,lsc*100)
    print("accuracy_score is:-",lss*100)
    print("\n")
```

# Observation: Here we can see that Accuracy score is same for all the croos validation values but the cross validation score is
# never same for any of them

# We can see that accuracy score and the Cross validation score are almost same ,this means that we are heading towards 
# a good model

```python
lsscore_selected = cross_val_score(lr,x,y,cv=6).mean()
print("The CV score is : ", lsscore_selected,"\nThe accuracy score is :", lss)
```
