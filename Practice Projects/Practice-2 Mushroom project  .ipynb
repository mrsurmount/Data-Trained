```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset1/master/mushrooms.csv")
```

```python
data
```

# Preprocessing:

#### Converting the variable in the numerical form, using LabelEncoder

```python
mappings = list()   
encoder = LabelEncoder()

for column in range(len(data.columns)):             
    data[data.columns[column]] = encoder.fit_transform(data[data.columns[column]])  
    mappings_dict = {index:label for index,label in enumerate(encoder.classes_)} 
    mappings.append(mappings_dict)
```

```python
mappings
```

```python
data
```

```python
y = data['class']
X = data.drop('class', axis = 1)
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.7)
```

# Model Selection:

```python
log_model = LogisticRegression()
svm_model = SVC(C=1.0, kernel = "rbf")  # in this svm model regularisation hyperparameter is taken as 1.0 and kernel is radiobasis function kernel
```

# Training:

#### Checking the dataset.

```python
np.sum(y)/len(y) 
```

```python
log_model.fit(X_train,y_train)
svm_model.fit(X_train,y_train)
```

```python
print(f"Logistic Regression: {log_model.score(X_test, y_test)}")
print(f"Support Vector Machine: {svm_model.score(X_test, y_test)}")
```
