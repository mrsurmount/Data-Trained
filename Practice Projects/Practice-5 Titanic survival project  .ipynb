# Handling the Missing values:

```python
# Dropping the column: "Cabin" from the dataset:
titanic_data = titanic_data.drop(columns = "Cabin", axis = 1)
titanic_data
```

## Observation:
### Here we can see that the column "Cabin" is dropped because that column has majority of null values and we cannot replace through any technique which is possible and not at all correct method so we dont have any other option except dropping the column, so we are dropping the column: "Cabin".

### Replacing the missing values in the column-"Age" through "mean":

```python
titanic_data["Age"].fillna(titanic_data["Age"].mean())
```

## Observation:
### Here we can see that we have replaced the column through "mean" of the column.

Here we will insert the Age replaced values into the dataset:

```python
titanic_data["Age"].fillna(titanic_data["Age"].mean(), inplace = True)
```

```python
# Checking the data info to see whether we handled the missing values in the column:"Age"
titanic_data.info()
```

## Observation:
### Here we can see that we have the missing values in the column: "Age".

### Replacing the missing values in the column-"Embarked" through "mode":

```python
# Finding the mode of the column: "Embarked":
print(titanic_data["Embarked"].mode())
```

## Observation: 
### Here we can see that the mode of the column "Embarked" is "S", now we will replace it into the dataset by using its "index".

```python
titanic_data["Embarked"].fillna(titanic_data["Embarked"].mode()[0], inplace = True)
```

```python
titanic_data.info()
```

## Converting the Categorical Columns into the Numerical Columns:

```python
titanic_data.replace({"Sex":{"male": 0, "female" : 1}, "Embarked" : {'S' : 0, 'Q' : 1, 'C' : 2}}, inplace = True)
```

```python
titanic_data.head()
```

## Separating the features and Target:

```python
x = titanic_data.drop(columns = ["PassengerId","Name","Ticket","Survived"], axis = 1)
y = titanic_data["Survived"]
```

```python
x
```

```python
y
```

# Splitting the data into training data and test data:

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
x.shape
```

# Model training:

## Logistic Regression:

```python
model = LogisticRegression()
```

```python
model
```

```python
# Training the model with training data:
model.fit(x_train,y_train)
```

```python
# Accuracy on the training data:
x_train_prediction = model.predict(x_train)
```

```python
x_train_prediction
```

```python
training_data_accuracy = accuracy_score(y_train,x_train_prediction)
```

```python
print("Accuracy Score of training data:" , training_data_accuracy)
```

## Observation:
### The accuracy score for our training data is 81.3%

```python
# Accuracy on the testing data:
x_test_prediction = model.predict(x_test)
```

```python
x_test_prediction
```

```python
testing_data_accuracy = accuracy_score(y_test,x_test_prediction)
```

```python
print("Accuracy Score of testing data:" , testing_data_accuracy)
```
